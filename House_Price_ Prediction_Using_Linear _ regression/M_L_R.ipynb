```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("C:/Users/sikan/Downloads/Log Regression and Clustering/Lab08/car selling dataset.csv")
```

```python
data.head(5)
```

```python
data.isnull().sum().sum()
```

```python
data=data.dropna()
```

```python
dummies=pd.get_dummies(data['name'])
dummies1=pd.get_dummies(data['fuel'])
dummies2=pd.get_dummies(data['owner'])
dummies3=pd.get_dummies(data['mileage'])
```

```python
dummies.head(5)
```

```python
merged=pd.concat([data,dummies,dummies1,dummies3],axis='columns')
```

```python
merged.head()
```

```python
final=merged.drop(['name','seller_type','transmission','torque','fuel','mileage','owner','engine','max_power'],axis='columns')
final.head()
```

```python
Y=final['selling_price']
Y.head()
```

```python
X=final.drop('selling_price',axis='columns')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
model= LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_prediction =  model.predict(x_test)
y_prediction
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
# predicting the accuracy score
score=r2_score(y_test,y_prediction)
print('r2 socre is ',score)
print('mean_sqrd_error is==',mean_squared_error(y_test,y_prediction))
print('root_mean_squared error of is==',np.sqrt(mean_squared_error(y_test,y_prediction)))
```

```python
model.score(x_test,y_test)
```
